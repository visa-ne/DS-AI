# Matplotlib tutorial

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
plt.plot([1,2,3,4], [10,200,300,400])
plt.show()
```

Now add labels and title to the above graph 

```python
plt.plot([1,2,3,4], [10,200,300,400])
plt.title(' Line chart')
plt.xlabel('Numbers')
plt.ylabel('Hunderds')
plt.show()
```

We can change the size of the figure by using the figure option 

```python
plt.figure(figsize = (20,10))
plt.plot([1,2,3,4], [10,200,300,400])
plt.show()
```

By default the a line chart with blue line (b-)
we can change it by passing a new argument alon with x and Y 

```python

plt.plot([1,2,3,4], [10,200,300,400], 'go') # go for green circles
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5, 5))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1],)
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```

```python
groups = ["A","B","C","D","E"]
indx = np.arange(len(groups))
sales = [100.0,220.0,150.0,180.0,230.0]
var = [10,9, 14, 15, 9]
type(sales)
type(groups)
plt.barh(indx, sales, xerr = var, color = "blue");
plt.yticks(indx, groups, fontsize=10, rotation=30);
#plt.show()

```

Scatter plot 

```python
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.scatter(x, y, color = "b")#, align='center')
plt.scatter(x2, y2, color='g')#, align='center')
plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
%matplotlib inline
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```
